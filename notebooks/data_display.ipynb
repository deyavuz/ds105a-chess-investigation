```python
import matplotlib.pyplot as plt
import numpy as np
```

Draft for the popularity line graphs

```python
x=[2020,2021,2022,2023,2024]
y=[4,5,6,7,8]
y1=[6,7,9,6,4]
name ="magnus"
country = 'Norway'
def create_line_graph_popularity(name,country,date,player_search,country_search):
    # date time values for trendline calculation and creating trendlines
    date_numeric = np.arctan(date)
    country_trend = np.poly1d(np.polyfit(date_numeric, country_search, 1))(date_numeric)
    player_trend = np.poly1d(np.polyfit(date_numeric,player_search,1))(date_numeric)

    #calculating the correlation
    correlation = np.corrcoef(country_search,player_search)[0,1]

    fig,ax1 = plt.subplots(figsize=(12,7))
    
    #plotting player search and trend line
    line1 = ax1.plot(date,player_search, marker='o', linestyle='-', color='black', label=f"{name}'s search")
    line2 = ax1.plot(date,player_trend, color = 'black', linestyle="--", label='player trendline')
    ax1.set_xlabel('Date', fontsize=12)
    ax1.set_ylabel(f"{name}'s search", color='black', fontsize=12)
    ax1.tick_params(axis='y', labelcolor='black')

    #plotting country search and trend line
    line3 = ax2 = ax1.twinx() #helps create a secondary y axis for country search values
    line4 = ax2.plot(date,country_search, marker='o', linestyle='-', color='red', label=f"{country}'s searc")
    ax2.plot(date,country_trend, linestyle='--',color='red',label="country trendline")
    ax2.set_ylabel('Country searched chess',fontsize=12, color='red')


    plt.title(f"{name}'s popularity graph ")
    plt.grid(alpha=0.1)

    plt.text(2020,8,f"Correlation: {correlation:.2f}", fontsize=12, bbox=dict(facecolor='white', alpha=0.5))
    #ax1.annotate('Major Tournament', xy=(2, 2300), xytext=(2, 2350),arrowprops=dict(facecolor='black', arrowstyle='->'),fontsize=10)

    lines = [line1, line2, line3, line4]
    labels = [f"{name}'s search","player trendline",f"{country}'s search","country trendline"]
    ax1.legend(handles = [line1, line2, line3, line4], labels = [[f"{name}'s search","player trendline",f"{country}'s search","country trendline"]], fontsize=10)
```

```python
create_line_graph_popularity(name,country,x,y,y1)
```
